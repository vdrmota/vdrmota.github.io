```python
import time

def make_request(url):
    return requests.get(url)
```

```python
api_results = pd.DataFrame()

c = 0
for index, row in df.iterrows():
        
    # get similar
    response = make_request("http://ws.audioscrobbler.com/2.0/?method=track.getsimilar&artist="+row['artist_name']+"&track="+row['track_name']+"&api_key=aafeb6ca46a3be3cd15a9bcad4e1066e&format=json")
    print(response.status_code)
    print(c)

    data = json.loads(response.text)
    print(data['similartracks'])
    dataframe = pd.DataFrame(data, index=[c])
    api_results = api_results.append(dataframe)
    c += 1
    break
```
